```python
import pandas as pd
```

```python
input_paths = ['data/raw/hungarian.data', 'data/raw/switzerland.data', 'data/raw/long-beach-va.data']

processed_data = []

cnt = 0

for inp in input_paths:
    with open(inp, 'r') as file:
        lines = file.readlines()

    flat_list = []
    for line in lines:
        values = line.split()
        for v in values:
            flat_list.append(v)

    # Convertimos la lista plana en líneas de longitud 76
    for i in range(0, len(flat_list), 76):
        line = flat_list[i:i+76]
        if len(line) == 76:  # Aseguramos que la longitud de la línea sea 76
            line[0] = str(cnt)
            processed_data.append(line)
            cnt += 1

# Convertimos los datos procesados en un DataFrame
df = pd.DataFrame(processed_data)

# Guardar el DataFrame combinado en un nuevo archivo CSV
df.head()
```
